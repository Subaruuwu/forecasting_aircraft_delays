!!!!ВСЕ ПРОПУСКИ ПОМЕЧАЮТСЯ КАК M!!!!

- station:
идентификатор погодной станции (в нашем случае KJFK)
- valid:
timestamp точки измерения, его можно перевести через pd.to_datetime
- tmpf:
температура в фаренгейтах на уровне земли
- dwpf:
температура точки росы в фаренгейтах
- relh:
относительная влажность
- drct:
направление ветра с "настоящего" севера
- sknt:
скорость ветра в узлах
- p01i:
Часовое количество осадков за период от времени наблюдения до времени предыдущего часового сброса осадков. В зависимости от местности этот показатель несколько различается. Значения приведены в дюймах. Это значение может содержать или не содержать замерзшие осадки, растопленные каким-либо устройством на датчике или оцененные каким-либо другим способом. К сожалению, нам не известна авторитетная база данных, указывающая, на какой станции установлен тот или иной датчик.
- alti:
настройки альтиметра (дюймы рт. ст. давления)
- mslp:
давление на уровне моря в миллибарах
- vsby:
видимость в милях
- gust:
порывы ветра в узлах
- skyc1:
описание облачности (на уровне 1)
- skyc2:
описание облачности (на уровне 2)
- skyc3:
описание облачности (на уровне 3)
- skyc4:
описание облачности (на уровне 4)
- skyl1:
высота уровня 1 в футах
- skyl2:
высота уровня 2 в футах
- skyl3:
высота уровня 3 в футах
- skyl4:
высота уровня 4 в футах
- wxcodes:
коды погоды разделенные через пробел https://www.aviation.govt.nz/assets/licensing-and-certification/meteorology/aerodrome-weather-codes.pdf 
- feel:
температура по ощущениям
- ice_accretion_1hr:
нарастание льда за 1 час (дюймы)
- ice_accretion_3hr:
нарастание льда за 3 час (дюймы)
- ice_accretion_6hr:
нарастание льда за 6 час (дюймы)


- ремарки:

- peak_wind_gust:
пиковые порывы ветра в узлах
- peak_wind_drct:
направление пикового порыва ветра в градусах
- peak_wind_time:
время пикового времени порыва
- metar:
необработанный метар

```python
data_enhanced = data[['station', 'valid', 'skyc1', 'skyc2', 'skyc3', 'skyc4', 'skyl1', 'skyl2', 'skyl3', 'skyl4']].copy()
```

```python
data['skyc4'].unique()
```

```python
skyc_dict = {
    'CLR': 0,
    'FEW': 1,
    'SCT': 2,
    'BKN': 3,
    'OVC': 4,
    'VV ': 5,
    'M': 0,
    0:0
}
```

```python
for col in ['skyc1', 'skyc2', 'skyc3', 'skyc4', 'skyl1', 'skyl2', 'skyl3', 'skyl4']:
    data_enhanced[col] = data_enhanced[col].astype(str).replace('M', 0)

for col in ['skyl1', 'skyl2', 'skyl3', 'skyl4']:
    data_enhanced[col] = data_enhanced[col].apply(lambda x: float(x) / 1000)

for col in ['skyc1', 'skyc2', 'skyc3', 'skyc4']:
    data_enhanced[col] = data_enhanced[col].apply(lambda x: skyc_dict[x])

for i, cols in enumerate(zip(['skyc1', 'skyc2', 'skyc3', 'skyc4'], ['skyl1', 'skyl2', 'skyl3', 'skyl4'])):
    col_1, col_2 = cols
    data_enhanced[f'inx_danjerous_{i + 1}'] = data_enhanced[col_1] / data_enhanced[col_2]
    data_enhanced[f'inx_danjerous_{i + 1}'] = data_enhanced[f'inx_danjerous_{i + 1}'].fillna(0)
    data_enhanced[f'inx_danjerous_{i + 1}'] = data_enhanced[f'inx_danjerous_{i + 1}'].apply(lambda x: 0 if x == np.inf else x)
```

```python
data_enhanced
```

```python
data_enhanced = convert_utc_to_jfk(data_enhanced)
```

```python
data_enhanced
```

```python
columns = ['valid', 'inx_danjerous_1', 'inx_danjerous_2', 'inx_danjerous_3', 'inx_danjerous_4', 'DATE_JFK', 'TIME_JFK']
data_to_merge = data_enhanced[columns].copy()
```

```python
data['valid'] = pd.to_datetime(data['valid'])
data_to_merge['valid'] = pd.to_datetime(data_to_merge['valid'])

data_with_dangerous = pd.merge(data, data_to_merge, how='left', on='valid')
```
